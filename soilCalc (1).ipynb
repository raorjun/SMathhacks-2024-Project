```python
def calcSolution(crop, liters):

    row = 0

    if crop == "cotton":
        row = 0
    elif crop == "rice":
        row = 1
    elif crop == "sugarcane":
        row = 2
    elif crop == "wheat":
        row = 3
    elif crop == "barley":
        row = 4
    elif crop == "legumes":
        row = 5
    elif crop == "sorghum":
        row = 6
    elif crop == "leafy vegetables":
        row = 7
    elif crop == "tomatoes":
        row = 8
    elif crop == "cucumbers":
        row = 9
    dfRow = data.iloc[row]
    return "To plant " + crop + " you will need " + str(dfRow.iloc[13]*liters) + " mg of Calcium nitrate, " + str(dfRow.iloc[14]*liters) + " mg of Potassium phosphate, " + str(dfRow.iloc[15]*liters) + " mg of Potassium nitrate, " + str(dfRow.iloc[16]*liters) + " mg of Magnesium sulfate, " + str(dfRow.iloc[17]*liters) + " mg of Chelated iron, " + str(dfRow.iloc[18]*liters) + " mg of Boron, " + str(dfRow.iloc[19]*liters) + " mg of Copper sulfate, " + str(dfRow.iloc[20]*liters) + " mg of Manganese sulfate, " + str(dfRow.iloc[21]*liters) + " mg of Zinc sulfate, and " + str(dfRow.iloc[22]*liters) + " mg of Sodium molybdate."
```

```python
calcSolution("cotton", 1)
```
